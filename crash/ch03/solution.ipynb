### 3.1 列表及其访问

1. 形如[1, 2, 3]
2. 使用过下标访问，从0开始或者从-1开始（倒序）开始


```python
# 3-1

names = ["aqiang", "dapang", "houzii"]
print(names[0])
print(names[1])
print(names[2])
```

```python
# 3-2

names = ["aqiang", "dapang", "houzii"]
print(names[0].title() + ", how are you?")
print(names[1].title() + ", how are you?")
print(names[2].title() + ", how are you?")
```

```python
# 3-3

moving = ["walking", "driving"]
print("I like go out by " + moving[0])
print("I like go out by " + moving[1])
```
