### 3.3 组织列表

1. 永久排序：sort()
2. 暂时排序：sorted()
3. 永久反序：reverse()
4. 列表长度：len()

```python
# 3-8

travels = ["江西", "新疆", "香港", "哈尔滨", "西安"]
# travels = ["jiangxi", "xinjiang", "xianggang", "haerbing", "xian"]

print(travels)
print(sorted(travels))
print(travels)
print(sorted(travels, reverse = True))
print(travels)
travels.reverse()
print(travels)
travels.reverse()
print(travels)
travels.sort()
print(travels)
travels.sort(reverse=True)
print(travels)
```
