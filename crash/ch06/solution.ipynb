```python
# 6-2

fn = {"mike": 1, "chanlder": 2, "zhuliie": 5, "justin": 7, "lora": 5}

for item in fn:
    print((item, fn[item]))
```

```python
# 6-3

program = {"for": "循环", "list": "列表"}

for item in program:
    print((item, program[item]))
```

```python
# 6-4

for k, v in program.items():
    print("key is " + str(k))
    print("value is " + str(v))
```
