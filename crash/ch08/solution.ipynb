```python
# 8-1 定义函数，调用函数

def display_message():
    print("I learn define function in this chapter.")

display_message()
```

```python
# 8-2 给函数穿参

def favorite_book(title):
    print("One of my favorite books is " + title)

favorite_book("Alice in Wonderland")
```

```python
# 8-3 按序传参，按关键字传参

def make_shirt(size, code):
    print("this shirt is " + size + " and with " + code)

make_shirt("XL", "Z")
make_shirt(size="L", code="A")
```

```python
# 8-4 默认参数

def make_shirt(size, code="I love Python"):
    print("this shirt is " + size + " and with " + code)

make_shirt("S")
```

```python
# 8-5 默认参数被传参

def describe_city(city="nanjing", country="China"):
    print(city + " is in " + country)

describe_city()
describe_city("wuhu", "china")
describe_city("paris", "franch")
```

```python
# 8-6 函数返回值

def city_country(city, country):
    return city + ", " + country

print(city_country("wuhu", "china"))
```

```python
# 8-7 返回高级数据类型

def make_album(name, album):
    return {"name": name, "album":album}

ab = make_album("zhoujielun", "qilixiang")
ab
```

```python
# 8-9 8-10 8-11 传递高级数据类型参数

def show_magicians(magicians):
    l = []
    for m in magicians:
        print(m)
        l.append("the Great " + m)
    return l

l = show_magicians(['ni', 'ta', 'wo'])
print(l)
```

```python
# 8-12 多参数传递

def make_sanwich(*shicai):
    for sc in shicai:
        print(sc)
make_sanwich('niurou', 'shucai', 'nainao')
```

```python
# 8-14

def make_car(maker, type, **kv):
    dic = {}
    dic['maker'] = maker
    dic['type'] = type
    for k, v in kv.items():
        dic[k] = v
    return dic
car = make_car('subaru', 'outback', color='blue', tow_package=True)
print(car)
```
